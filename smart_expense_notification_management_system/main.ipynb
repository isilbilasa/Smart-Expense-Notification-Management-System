## Kullanılan Teknik Konular:

Proje, Python'ın ileri seviye yeteneklerini eksiksiz bir şekilde harmanlamaktadır:


**Modüler Mimari:** Kodun; models, notifications ve utils olarak parçalanması (Separation of Concerns).

Nesne Yönelimli Programlama (OOP):

**Inheritance:** Bildirim sınıflarının tek bir merkezden türetilmesi.

**Polymorphism:** send() metodunun Email ve SMS sınıflarında farklı davranışlar sergilemesi.

**Özel Metotlar:** __str__ ile nesne okunabilirliği ve sum/len ile koleksiyon yönetimi.

**Hata Yönetimi (Error Handling):** try-except blokları ile kullanıcı hatalarına karşı dayanıklı (robust) veri girişi.

**Veri Yapıları:** Dinamik veri yönetimi için List, benzersiz kategori analizi için Set kullanımı.

**Standart Kütüphaneler:** Zaman yönetimi için datetime ve dinamik simülasyonlar için random entegrasyonu.

## Neden Gerçekçi?

Bu sistem, günümüzdeki FinTech (Finansal Teknolojiler) uygulamalarının temel çalışma mantığını simüle eder:

**Dinamik Veri Akışı:** Statik veriler yerine kullanıcıdan alınan canlı verilerle işlem yapar.

**Otomasyon:** Manuel kontrol yerine, sistemin bütçe durumunu otonom olarak hesaplayıp karar vermesi (Limit aşımı kontrolü).

**Çok Kanallı Bildirim (Omnichannel):** Gerçek dünyadaki gibi, sistemin altyapısını değiştirmeden yeni bildirim kanallarının (WhatsApp, Slack vb.) eklenebilmesine olanak tanıyan esnek bir mimariye sahiptir.

```python
from models import Budget
from utils import get_user_expenses

my_budget = Budget(limit=5000)

user_data = get_user_expenses()

for exp in user_data:
    my_budget.add_expense(exp)

print(f"Recorded expense count: {my_budget.expense_count()}")
```

```python
print("\n--- Expense List ---")
for exp in my_budget.expenses:
    print(exp)

total = my_budget.total_expense()
average = my_budget.average_expense()
max_exp = my_budget.max_expense()
min_exp = my_budget.min_expense()
categories = my_budget.category_set()

print("\n--- Budget Analysis ---")
print(f"Total expense: {total:.2f} TL")
print(f"Average expense: {average:.2f} TL")
print(f"Highest expense: {max_exp}")
print(f"Lowest expense: {min_exp}")
print(f"Categories: {categories}")
```

```python
from notifications import EmailNotification, SMSNotification
from utils import generate_random_notification, register_notifications

notifications = []

if my_budget.is_limit_exceeded():
    alert_message = f"Budget limit exceeded! Limit: {my_budget.limit:.2f} TL | Current total: {my_budget.total_expense():.2f} TL"
    notifications.append(EmailNotification(alert_message))
    notifications.append(SMSNotification(alert_message))
    notifications.append(generate_random_notification("Automatic finance alert created."))
else:
    info_message = f"Budget is under control. Remaining amount: {my_budget.limit - my_budget.total_expense():.2f} TL"
    notifications.append(EmailNotification(info_message))
    notifications.append(generate_random_notification("Daily finance status created."))

sent_notifications = register_notifications(*notifications)

print("\n--- Sent Notifications ---")
for item in sent_notifications:
    print(item)
```

```python
print("\n--- Final Summary ---")
print(f"Budget limit: {my_budget.limit:.2f} TL")
print(f"Total expense: {my_budget.total_expense():.2f} TL")
print(f"Average expense: {my_budget.average_expense():.2f} TL")
print(f"Expense count: {my_budget.expense_count()}")
print(f"Limit status: {'Exceeded' if my_budget.is_limit_exceeded() else 'Not Exceeded'}")
print(f"Unique category count: {len(my_budget.category_set())}")
```
